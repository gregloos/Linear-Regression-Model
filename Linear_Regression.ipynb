# In this file i will make a linear regression model from scratch

```python
import pandas as pd # used to import a dataset to test the model
import matplotlib.pyplot as plt # used to visualize the data
```

```python
DataPath = "C:\\Users\\gregl\\Desktop\\ML_Projects\\Linear Regression Model\\simple_linear_data.csv"
data = pd.read_csv(DataPath)    
```

```python
data.head()
```

```python
plt.scatter(data.feature, data.target)
plt.show()
```

```python
def gradient_descent(m_now, b_now, points, L):
    m_gradient = 0
    b_gradient = 0
    n = len(points)
    
    for i in range(n):
        x = points.iloc[i].feature
        y = points.iloc[i].target
        
        m_gradient += -(2/n) * x * (y - (m_now * x + b_now))   
        b_gradient += -(2/n) * (y - (m_now * x + b_now)) 
    m = m_now - m_gradient * L
    b = b_now - b_gradient * L
    return m, b

m = 0
b = 0
L = 0.001
epochs = 1000
```

```python
for i in range(epochs):
    if i % 50 == 0:
        print(f"Epoch {i}")
    m, b = gradient_descent(m, b, data, L)
print(m, b)
```

```python
plt.scatter(data.feature, data.target, color="black")
plt.plot(list(range( -10, 15)), [m * x + b for x in range(-10, 15)], color="red")
plt.show()
```
